**About MobileNetV2**

MobileNetV2 stands as a convolutional neural network (CNN) design crafted to excel in efficient and lightweight tasks like image classification and feature extraction. It emerges as the successor to the initial MobileNet architecture, emerging from the work of Google researchers. The primary goal was to meet the demand for deep learning models capable of functioning optimally on devices like mobile phones and embedded systems, which often have constraints on computational capabilities. This achievement is realized through the strategic application of two techniques: depthwise separable convolutions and inverted residual blocks.


*   **Depthwise Separable Convolutions:**
MobileNetV2 relies heavily on depthwise separable convolutions, a technique that divides a standard convolution into two stages: depthwise convolution, where filters are applied individually to each input channel, followed by pointwise convolution, which employs 1x1 convolutions to merge the output channels. This approach effectively reduces the computational load when compared to conventional convolutions.

*   **Inverted Residual Blocks**:
MobileNetV2 introduces inverted residual blocks, often referred to as bottleneck blocks. These blocks encompass the following components:


> 1. A pointwise convolution to diminish the number of input channels.
> 1. A depthwise separable convolution to extract features.
> 3. Another pointwise convolution to augment the count of output channels.
> 4. A potential skip connection (activated when channel numbers align) to uphold vital features.

**STEP 2 - Library Import**

Import necessary libraries, including TensorFlow, MobileNetV2 model, preprocessing functions, and NumPy.

```python
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
```

**STEP 3 - Load pre-trained Model**

The MobileNetV2 model with pre-trained weights on the ImageNet dataset is loaded.

```python
model = MobileNetV2(weights='imagenet')
```

**STEP 4 - Load and Prepare the Image**

In this code, I have uploaded an image as 'GOKU.jpeg', and then resized to match MobileNetV2's input size (224x224), and converted it to a preprocessed NumPy array.

```python
image_path = 'GOKU.jpeg'
img = image.load_img(image_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)
```

**STEP 5 - Display the Image**

To check if image has been loaded correctly

```python
from IPython.display import Image, display
display(Image(image_path,height=300,width=400))
```

**STEP 6 - Make Prediction**

The MobileNetV2 model predicts class probabilities for the input image.

```python
predictions = model.predict(img_array)
decoded_predictions = decode_predictions(predictions, top=3)[0]

for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
    print(f"{i + 1}: {label} ({score:.2f})")
```

# Why do the size of images have to be 224x224 pixels?

The size of 224x224 pixels is a common input size requirement for many pre-trained convolutional neural network (CNN) architectures, including MobileNetV2. This specific input size is used for a few reasons:


*   **Consistency:** Pre-trained models have been trained on large datasets with images of a specific size, and using the same input size ensures compatibility with the model's architecture and weights.

*   **Memory and Computational Efficiency:** Using a consistent input size helps optimize memory usage and computational efficiency during training and inference. It also enables batch processing of images.

*   **Pooling Layers:** CNN architectures often include pooling layers, which reduce spatial dimensions. Consistent input sizes make it easier to determine the behavior and effects of these layers.

When using pre-trained models like MobileNetV2, it's essential to match the input size to what the model expects to achieve accurate results and avoid errors.
